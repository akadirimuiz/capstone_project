```python
# creating split of train_test of the dataset
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size =0.2)
```

```python
# i want to train classification model with  Decision tree clasiffier
#import the classification library
from sklearn.tree import DecisionTreeClassifier
```

```python
#initialize the model
Dec_model = DecisionTreeClassifier()
```

```python
# fit in the model
Dec_model = Dec_model.fit(x_train,y_train)
```

```python
Dt_preds = Dec_model.predict(x_test)
```

```python
#classification report
from sklearn.metrics import classification_report
print(classification_report(Dt_preds,y_test))
```
